```python
import numpy as np
from sklearn import preprocessing

sample_data = np.array([[2.1, -1.9, 5.5],
                        [-1.5, 2.4, 3.5],
                        [0.5, -7.9, 5.6],
                        [5.9, 2.3, -5.8]])

sample_data
```

```python
sample_data.shape
```

Teknik Binarisation

```python
sample_data
```

```python
# parameter yang digunakan untuk ambang batas
preprocessor = preprocessing.Binarizer(threshold=0.5)
binarised_data = preprocessor.transform(sample_data)
binarised_data
```

Teknik Scalling

```python
sample_data
```

```python
# Scalling Min Max Scaller
preprocessor = preprocessing.MinMaxScaler(feature_range=(0,0.1))
preprocessor.fit(sample_data)

#Cara 1
scaled_data = preprocessor.transform(sample_data)
scaled_data
```

```python
#Cara 2
scaled_data = preprocessor.fit_transform(sample_data)
scaled_data
```

Teknik Normalization L1

```python
sample_data
```

```python
l1_normalizer = preprocessing.Normalizer(norm='l1')
l1_normalized_data = l1_normalizer.fit_transform(sample_data)
l1_normalized_data
```

Teknik Normalization L2

```python
sample_data
```

```python
l1_normalizer = preprocessing.Normalizer(norm='l2')
l1_normalized_data = l1_normalizer.fit_transform(sample_data)
l1_normalized_data
```

Encoding

```python
label_kategori = ['senin', 'selasa', 'rabu', 'kamis', 'jumat', 'sabtu', 'minggu']
encoder = preprocessing.LabelEncoder()
encoder.fit(label_kategori)

print("\nLabel mapping:")
for i, item in enumerate(encoder.classes_):
    print(item, '=', i)
```

Latihan Cross Validation

```python
import sklearn
from sklearn import datasets

#Load iris dataset
iris = datasets.load_iris()
```

```python
# mendefinisikan atribut dan label pada dataset
x=iris.data
y=iris.target
```

```python
from sklearn import tree

# membuat model dengan decision tree classifier
clf = tree.DecisionTreeClassifier()
```

```python
from sklearn.model_selection import cross_val_score

# mengevaluasi performa model dengan cross_val_score
scores = cross_val_score(clf, x, y, cv=5)
```
